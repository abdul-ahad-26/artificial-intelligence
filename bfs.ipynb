```python
from collections import deque
```

```python
def bfs(graph, start, goal):
    queue = deque([[start]]) # queue of path
    visited = set([start]) 


    while queue:
        path = queue.popleft()
        vertex = path[-1]
        if vertex == goal:
             return path
             
        for neighbour in graph[vertex]:
                if neighbour not in visited:
                    visited.add(neighbour)
                    new_path = path + [neighbour]
                    queue.append(new_path)
    return None
```

```python
graph = {0: [1,2,3], 1 : [0,2], 2:[0, 1, 4 ], 3 : [0], 4: [2]}
start = 0
goal = 4
shortest_path = bfs(graph, start, goal)
print("Shortest path: ", shortest_path)
```

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()

# Create grapgh
for node, neighbours in graph.items():
    for neighbour in neighbours:
        G.add_edge(node, neighbour)
# Shortest path edges
path_edges = list(zip(shortest_path, shortest_path[1:]))

pos = nx.spring_layout(G)# position layout for better visuals



nx.draw(G, pos, with_labels= True, node_color='lightblue', edge_color='gray', node_size=700 )
nx.draw_networkx_edges(G, pos, edgelist=path_edges, edge_color='red', width=2)
nx.draw_networkx_nodes(G, pos, nodelist=shortest_path, node_color='red')

plt.title(f"Shortest Path from {start} to {goal}: {shortest_path}")
plt.show()
```
