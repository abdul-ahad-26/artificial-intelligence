## TASK 2: Data Cleaning


```python
df.drop_duplicates(inplace=True)
df
```

```python
IMDB_Rating_mean = df['IMDB_Rating'].mean()
df['IMDB_Rating'].fillna(IMDB_Rating_mean)
df['IMDB_Rating']
```

```python
df['Runtime'] = df['Runtime'].str.replace('min','').astype(int)
```

```python
df['Runtime']
```

## Task 3: Data Filtering & Sorting 


### 1. Find all movies with an IMDb rating greater than 8.5. 


```python
df[df['IMDB_Rating'] > 8.5]
```

### 2. List movies that belong to the Action or Sci-Fi genre. 


```python
df[df['Genre'].str.contains('Action | Sci-Fi') ]
```

### 3. Find movies that were released between 2000 and 2015. 


```python
df[( df['Released_Year'] < "2015") & ( df['Released_Year'] > "2000")]
```

### 4. Sort the dataset based on IMDb rating in descending order. 

```python
df.sort_values(by="IMDB_Rating",ascending=False, inplace=True)
df
```

## Data Aggregation & Grouping 

### 1. Find the average IMDb rating for each genre. 


```python
# Split genre string into list
df["Genre"] = df["Genre"].str.split(", ")

# Explode to get one genre per row
df = df.explode("Genre")

# Group by individual genres and calculate average rating
average_ratings = df.groupby("Genre")["IMDB_Rating"].mean()

average_ratings
```

### 2. Determine which year had the most movies released.

```python
# per_year_releases = df.groupby("Released_Year")["Series_Title"].count()
# per_year_releases [per_year_releases == per_year_releases.max()]
top_year = df["Released_Year"].value_counts().idxmax()
print("Most movie released year: ", top_year)
```

### 3. Find the top 5 directors who have directed the most movies in the dataset. 

```python
top_five_directors = df["Director"].value_counts().head(5).index
print("Top five directors: ", top_five_directors)
```

## Visualization 
Matplotlib or Seaborn 

### 1. Plot a histogram of IMDb ratings. 

```python
import matplotlib.pyplot as plt
ratings = df["IMDB_Rating"]
plt.hist(ratings,bins=10, color='skyblue', edgecolor='black')
plt.xlabel("IMDB Rating")
plt.ylabel("Number of Movies")
plt.title("Distributions of IMDB Ratings")
plt.show()
```

### 2. Create a bar chart showing the top 10 genres with the most movies. 


```python
top_genres = df["Genre"].value_counts().head(10)
plt.bar(top_genres.index, top_genres.values)
plt.xticks(rotation=45)
plt.title("Top 10 Genres by Count")
plt.ylabel("Number of Movies")
plt.show()
```

### 3. Visualize the trend of IMDb ratings over the years. 

```python
avg_ratings_per_year = df.groupby('Released_Year')["IMDB_Rating"].mean()
avg_ratings_per_year.plot()
plt.xlabel("Year")
plt.ylabel("Average IMDb Rating")
plt.title("IMDb Ratings by Year")
plt.show()
```
